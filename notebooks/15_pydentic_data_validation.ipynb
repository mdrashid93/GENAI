```python
from pydantic import BaseModel,Field
from typing import List,Annotated
```

```python
class User(BaseModel):
    name:str
    age:int
```

```python
user=User(name="md", age=22)
user
```

Field validation

```python
class Product(BaseModel):
    title:str =Field(max_length=10)
    des=str=Field(min_length=20)
    price:int = Field(gt=0,le=100)
```

```python
product=Product(title="aisa", des="thats cute girl in the world", price=1)
product2=Product(title="Aisa", des="that cute girl in the world", price=1)
```

```python
class Order(BaseModel):
    id:str=Field(min_length=6, max_length=12)
    product:Product
```

```python
order=Order(id="apple", product=product)
```

```python
order.product.price
```

```python
class Carts(BaseModel):
    id:str=Field(min_length=6, max_length=12)
    items: List[Product]
    comment:str=Field(
        default="no comment"
    )
```

```python
cart=Carts(id="aaaa", items=[product,product2], comment="this is show")
```

```python
cart
```

```python
for product in cart.items:
    print(product.title)
```

Annotated

```python
common_validator=Annotated[List[Product], Field(default=[])]
```

```python
class Carts(BaseModel):
    id:str =Field(min_length=6, max_length=12)
    items:common_validator
    comment:str=Field(
        default="No Comment"
    )
```

```python
class Order(BaseModel):
    products:common_validator
```

```python
ord=Order(products=[product,product2])
```
