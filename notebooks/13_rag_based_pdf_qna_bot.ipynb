```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings,ChatOpenAI
from langchain_core.prompts import PromptTemplate
```

```python
loader=PyPDFLoader("data/sample.pdf")
docs=loader.load()
len(docs)
```

```python
splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
splitted_data=splitter.split_documents(docs)
len(splitted_data)
```

```python
embeddings=OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
vector_store=Chroma.from_documents(
    documents=splitted_data,
    embedding=embeddings)
```

```python
query="what is machine learning datascience?"
data=vector_store.similarity_search(query=query , k=2)
```

```python
context=""
for doc in data:
    context +=doc.page_content + "\n"
print(context)
```

```python
llm=ChatOpenAI(model="gpt-5-")
res=llm.invoke(f"""can you provide answer based on
               provided context for my question, context:{context} and question:{query}""")
```

```python
print(res.content)
```

chain context_generate | prompt | llm | strparser

```python
def get_context(query:str):
    data=vector_store.similarity_search(query=query )
    context=""
    for doc in data:
        context +=doc.page_content + "\n"
    return  {"context":context,
             "query":query}
    
```

```python
prompt=PromptTemplate.from_template("""
you a helpful assistant who provides answer based on provided context for my question,
if you dont know the answer based on provided context then say you dont know,
context:{context} and 
question:{query}
    """)
```

```python
rag=_chain=get_context | prompt | llm
```

```python
res=rag.invoke("what is duration of machine learning course?")
```

```python
print(res.content)
```
