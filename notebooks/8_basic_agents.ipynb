```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int, b:int):
    """
    it will return the sum of tow numbers

    Args:
        a (int): _description_
        b (int): _description_
    """
    return a+b

@tool
def multiply_numbers(a:int, b:int):
    """
    it will return the multiply of tow numbers

    Args:
        a (int): _description_
        b (int): _description_
    """
    return a*b


# def substract_numbers(a:int, b:int):
#     """
#     it will return the substract of tow numbers

#     Args:
#         a (int): _description_
#         b (int): _description_
#     """
#     return a-b
```

```python
add_numbers.invoke({"a":11,"b":4})
```

```python
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent
```

```python
llm=ChatOpenAI(model="gpt-4")
```

```python
agent=create_agent(
    model=llm,
    tools=[add_numbers,multiply_numbers ],
    system_prompt="you are math teacher , and always use tool for calculation."
)
```

```python
response=agent.invoke({"messages"[{"role":"user","content":"what is 2+2*2 ?"}]})
print(response["messages"][-1].content)
```
