```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
text="hello , my name is md_rashid, email is md.rasshid9@gmail.com , my age is 22"
```

```python
res=llm.invoke(f"please give me only name, email and age from this text: {text}")
```

```python
print(res.content)
```

Solution structure output

```python
from pydantic import BaseModel,Field
from typing import List

class ResponseStrecture(BaseModel):
    name:str=Field(description="name of the user")
    email:str=Field(description="email address")
    age:int=Field(description="age")
strectured_llm=llm.with_structured_output(ResponseStrecture)
    
```

```python
res=strectured_llm.invoke(f"please give me only name, email and age from this text: {text}")
```

```python
res.model_dump()
```
