Search similarity

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
documents=[
           "machine learning is subset of artificial intelligence",
           "deep learning is uses neural network with multiple layers",
           "python is a programming languages popular for datascience",
           "india won cricket world cup in 2026",
           "AI models are trained on large datasets to learn patterns and make predictions"
           ]
```

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings=OpenAIEmbeddings(model="text-embedding-3-small", dimensions=1024)
```

```python
vector=embeddings.embed_documents(documents)
```

```python
len(vector[0])
```

Store vector embedding in vector db

```python
from langchain_chroma import Chroma
```

```python
vector_store=Chroma.from_texts(texts=documents,
                               embeddings=embeddings,
                               persist_directory="./vector_db")
```

```python
query="what is machine learning"
resutl=vector_store.similarity_search(query=query , k=2)
resutl
```
